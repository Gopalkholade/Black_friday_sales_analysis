```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('BlackFriday.csv')
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
df.groupby('Age').size().plot(kind='bar',figsize=(12,6),title='Purchase distributioon by age')
```

```python
for i in df['Age'].unique():
    print([i,':',len(df[df['Age']==i]),':',df[df['Age']==i]['Product_ID'].nunique()])
```

```python
lst=[]
for i in df['Age'].unique():
    lst.append([i,df[df['Age']==i]['Product_ID'].nunique()])
```

```python
data=pd.DataFrame(lst,columns=['Age','Product'])
```

```python
data.plot.bar(x="Age",figsize=(8,6))
```

```python
df.groupby('Age').sum().plot(kind='bar',figsize=(12,6),title='Amount spent by age')
```

```python
df.groupby('Age').mean()['Purchase'].plot(kind='pie',autopct='%.3f')
```

```python
df.groupby('Marital_Status').size().plot(kind='pie',autopct='%.3f')
```

```python
df.groupby('Gender').size().plot(kind='pie',autopct='%.3f')
```
